```python
import stanza
from tqdm import tqdm

stanza.download('en', package='mimic', processors={'ner': 'i2b2'})
nlp = stanza.Pipeline('en', package='mimic', processors={'ner': 'i2b2'})

vaccine_adverse_affects = [[] for x in range(len(filtered_VAERSDATA_df["SYMPTOM_TEXT"]))]
for i, symptom_description in tqdm(enumerate(filtered_VAERSDATA_df["SYMPTOM_TEXT"])):
    output = nlp(symptom_description)
    for ent in output.entities:
        if ent.type == "PROBLEM":
            vaccine_adverse_affects[i].append(ent.text)
#         print(f'{ent.text}\t{ent.type}')
```

```python
vaccine_adverse_affects = [vaccine_adverse_affect for vaccine_adverse_affect in vaccine_adverse_affects if vaccine_adverse_affect]
vaccine_adverse_affects = vaccine_adverse_affects[0:20]
print(len(vaccine_adverse_affects))
```

# Step2: Linking to Standard Terms

### Using Exact match

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

def remove_stopwords(word_list):
    stop_words = set(stopwords.words('english'))  # You can change 'english' to another language if needed
    filtered_words = [word.lower() for word in word_list if word.lower() not in stop_words]
    return filtered_words

most_symp_list = remove_stopwords(most_symp.keys())


std_symptoms = [[0] * len(vaccine_adverse_affect) for vaccine_adverse_affect in vaccine_adverse_affects]
for i, vaccine_adverse_affect in enumerate(vaccine_adverse_affects):
    for j, symptom in enumerate(vaccine_adverse_affect):
        if symptom.lower() in most_symp_list:
            std_symptoms[i][j] = symptom

exact_match_symptoms = std_symptoms
```

```python
i=0
for model_out, std_sym in zip(vaccine_adverse_affects[i], std_symptoms[i]):
    print("model_output:", model_out, ":::: ", "standard_term:", std_sym)
```

### Using Fuzzy matching

```python
from fuzzywuzzy import fuzz
# Create a list of lists with all values equal to 0
best_score = [[0] * len(vaccine_adverse_affect) for vaccine_adverse_affect in vaccine_adverse_affects]
std_symptoms = [[0] * len(vaccine_adverse_affect) for vaccine_adverse_affect in vaccine_adverse_affects]
for i, vaccine_adverse_affect in enumerate(vaccine_adverse_affects):
    for j, symptom in enumerate(vaccine_adverse_affect):
        for symptom_standard in most_symp.keys():
            similarity_score = fuzz.token_set_ratio(symptom, symptom_standard)
            if similarity_score > best_score[i][j]:
                best_score[i][j] = similarity_score
                std_symptoms[i][j] = symptom_standard

fuzzy_match_symptoms = std_symptoms
```

```python
i=0
for model_out, std_sym in zip(vaccine_adverse_affects[i], std_symptoms[i]):
    print("model_output:", model_out, ":::: ", "standard_term:", std_sym)
```

### Using word embedding similarity

```python
from transformers import BertTokenizer, BertModel, AutoTokenizer, AutoModel
import torch
```

```python
# model_name = 'bert-base-uncased'
# tokenizer = BertTokenizer.from_pretrained(model_name)
# model = BertModel.from_pretrained(model_name)

# model_name = 'monologg/biobert_v1.1_pubmed'
# tokenizer = AutoTokenizer.from_pretrained(model_name)
# model = AutoModel.from_pretrained(model_name)

# Load clinical BERT model and tokenizer
model_name = "emilyalsentzer/Bio_ClinicalBERT"
model = BertModel.from_pretrained(model_name)
tokenizer = BertTokenizer.from_pretrained(model_name)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def get_embedding(text, tokenizer, model):

    # Tokenize input text
    tokens = tokenizer(text, return_tensors='pt')

    # Obtain embeddings
    with torch.no_grad():
        outputs = model(**tokens)
        embeddings = outputs.last_hidden_state

    # Mean pooling
    mean_pooled = embeddings.mean(dim=1)
    return mean_pooled
```

```python
standard_symptoms = []
for symptom in most_symp.keys():
    standard_symptoms.append(get_embedding(symptom, tokenizer, model))
standard_symptoms = torch.concat(standard_symptoms)
print(standard_symptoms.shape)
```

```python
# Create a list of lists with all values equal to 0
best_score = [[0] * len(vaccine_adverse_affect) for vaccine_adverse_affect in vaccine_adverse_affects]
std_symptoms = [[0] * len(vaccine_adverse_affect) for vaccine_adverse_affect in vaccine_adverse_affects]
for i, vaccine_adverse_affect in tqdm(enumerate(vaccine_adverse_affects)):
    for j, symptom in enumerate(vaccine_adverse_affect):
        for symptom_standard_text, symptom_standard in zip(most_symp.keys(), standard_symptoms):
            similarity_score = cosine_similarity(get_embedding(symptom, tokenizer, model), symptom_standard.unsqueeze(dim=0))
            if similarity_score > best_score[i][j]:
                best_score[i][j] = similarity_score
                std_symptoms[i][j] = symptom_standard_text

bert_match_symptoms = std_symptoms
```

# Evaluation

```python
from itertools import chain
output = list(chain(*exact_match_symptoms))
predictions = len(output)

tp = 0
fp = 0
fn = 0
for out in output:
    if out in most_symp.keys():
        tp+=1
    else:
        fp +=1

for sympt in most_symp.keys():
    if sympt not in output:
        fn+=1


precision = tp/(tp + fp)
recall = tp/(tp + fn)
f1_score = (2 * precision * recall)/(precision + recall)
print(f"The precision of model using exact matching is: {precision}")
print(f"The recall of model using exact matching is: {recall}")
print(f"The f1_score of model using exact matching is: {f1_score}")
```

```python
from itertools import chain
output = list(chain(*fuzzy_match_symptoms))
predictions = len(output)

tp = 0
fp = 0
fn = 0
for out in output:
    if out in most_symp.keys():
        tp+=1
    else:
        fp +=1

for sympt in most_symp.keys():
    if sympt not in output:
        fn+=1


precision = tp/(tp + fp)
recall = tp/(tp + fn)
f1_score = (2 * precision * recall)/(precision + recall)
print(f"The precision of model using fuzzy matching is: {precision}")
print(f"The recall of model using fuzzy matching is: {recall}")
print(f"The f1_score of model using fuzzy matching is: {f1_score}")
```

```python
from itertools import chain
bert_output = list(chain(*bert_match_symptoms))
predictions = len(bert_output)

tp = 0
fp = 0
fn = 0
for out in bert_output:
    if out in most_symp.keys():
        tp+=1
    else:
        fp +=1

for sympt in most_symp.keys():
    if sympt not in bert_output:
        fn+=1


precision = tp/(tp + fp)
recall = tp/(tp + fn)
f1_score = (2 * precision * recall)/(precision + recall)
print(f"The precision of model using embeddings is: {precision}")
print(f"The recall of model using embeddings is: {recall}")
print(f"The f1_score of model using embeddings is: {f1_score}")
```
